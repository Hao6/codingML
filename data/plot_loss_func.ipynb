```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def crossEntroyLoss(y, t=1):
    return -t*np.log(y)-(1-t)*np.log(1-y)
def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))
def squareLoss(y, t):
    return 1./2 * (y - t)**2
```

```python
y = np.linspace(0.0001, 0.9999, 500)
```

```python
plt.figure(figsize=(12, 7))
plt.plot(y, crossEntroyLoss(y, 1), color='blue', label='t=1', linewidth=3)
plt.plot(y, crossEntroyLoss(y, 0), color='lightgreen', label='t=0', linewidth=3)
plt.legend(loc=9)
plt.xlim((0, 1))
plt.ylim((0, 5))
plt.xlabel('y')
plt.ylabel('cross_entroy_loss')
plt.grid()
```

```python
x = np.linspace(-10, 10, 2000)
```

```python
plt.figure(figsize=(12, 7))
plt.plot(x, crossEntroyLoss(sigmoid(x), 1), color='blue', label='t=1', linewidth=3)
plt.plot(x, crossEntroyLoss(sigmoid(x), 0), color='lightgreen', label='t=0', linewidth=3)
plt.legend(loc=9)
# plt.xlim((0, 1))
# plt.ylim((0, 5))
# plt.axvline(x=0.9, ymin=0, ymax = crossEntroyLoss(0.9, t=0)/5, linewidth=2, color='k')
# plt.axvline(x=0.95, ymin=0, ymax = crossEntroyLoss(0.95, t=0)/5, linewidth=2, color='k')
# plt.axhline(y=crossEntroyLoss(0.9, t=0), xmin=0, xmax = 0.9, linewidth=2, color='k')
# plt.axhline(y=crossEntroyLoss(0.95, t=0), xmin=0, xmax = 0.95, linewidth=2, color='k')
plt.xlabel('x')
plt.ylabel('cross_entroy_loss')
plt.grid()
```

```python
plt.figure(figsize=(12, 7))
plt.plot(x, sigmoid(x), color='blue', label='sigmoid', linewidth=3)
plt.legend(loc=9)
plt.ylim((0, 1))
plt.axvline(x=0, c='k', linewidth=2)
plt.xlabel('x')
plt.ylabel('sigmoid')
plt.grid()
```
