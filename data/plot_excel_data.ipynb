```python
"""
# Definition for a QuadTree node.
class Node(object):
    def __init__(self, val, isLeaf, topLeft, topRight, bottomLeft, bottomRight):
        self.val = val
        self.isLeaf = isLeaf
        self.topLeft = topLeft
        self.topRight = topRight
        self.bottomLeft = bottomLeft
        self.bottomRight = bottomRight
"""
class Solution(object):
    def construct(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: Node
        """
        def help(grid, x_lower, y_lower, width):
            cur_val = grid[x_lower][y_lower]
            temp_root = Node(cur_val, True)
            if width > 1:
                for i in range(x_lower, x_lower+width):
                    for j in range(y_lower, y_lower+width):
                        if grid[i][j] != cur_val:
                            temp_root.isLeaf = False
                            temp_root.topLeft = help(grid, x_lower, y_lower, width/2)
                            temp_root.topRight = help(grid, x_lower, y_lower+width/2, width/2)
                            temp_root.bottomLeft = help(grid, x_lower + width/2, y_lower, width/2)
                            temp_root.bottomRight = help(grid, x_lower + width/2, y_lower + width/2, width/2)
                            i = width + 1
                            break
            return temp_root
        return help(grid, 0, 0, len(grid))
        
        
```
