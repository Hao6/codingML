```python
import matplotlib.pyplot as plt
import numpy as np
```

$$
f(x | \mu, b)=\frac{1}{2 b} \exp \left(-\frac{|x-\mu|}{b}\right)
$$

$$
f(x)=\frac{1}{\sigma \sqrt{2 \pi}} \exp{-\frac{(x-\mu)^{2}}{2 \sigma^{2}}}
$$

```python
def normal(x, u=0, delta=0.5):
    return 1.0/(delta*np.sqrt(2*np.pi))*np.exp(-np.square(x-u)/(2*delta**2))
def laplace(x, u=0, b=0.5):
    return 1.0/(2*b)*np.exp(-np.abs(x - u)/b)
```

```python
x = np.linspace(-5, 5, 1000)
```

```python
def plot_normal(x):
    plt.figure(figsize=(12, 7))
    plt.grid()
    deltas = [1, 2, 4]
    for delta in deltas:
        plt.plot(x, normal(x, delta=delta), label='u='+str(0)+', $\delta$='+str(delta))
    plt.ylim(0, 1)
    plt.legend()
```

```python
plot_normal(x)
```

```python
def plot_laplace(x):
    plt.figure(figsize=(12, 7))
    plt.grid()
    deltas = [1, 2, 4]
    for delta in deltas:
        plt.plot(x, laplace(x, b=delta), label='u='+str(0)+', b='+str(delta), alpha=0.7)
    plt.ylim(0, 1)
    plt.legend()
```

```python
plot_laplace(x)
```
