## 懒惰版knn
- 不用预先训练模型，直接进行比较得出结果
- 最大的问题就是计算太慢了，所以需要提前处理训练数据，使之能够被高效而不是线性的检索

```python
class Knn(object):
    def __init__(self, n_neighbours = 5, distance='European', verbose=False, 
                 classification =True):
        self.n_neighbours = n_neighbours
        self.distance = distance
        self.verbose = verbose
        self.train_x = None
        self.train_y = None
        self.classification = classification
    def train(self, train_x, train_y):
        self.train_x = train_x
        self.train_y = train_y
    def fit(self, train_x, train_y):
        self.train(train_x, train_y)
    def predict(self, test_x):
        pre = np.zeros(test_x.shape[0])
        for i in range(test_x.shape[0]):
            if self.verbose and (i+1) % 10 == 0:
                print 'the %d test datas is predicting' % (i+1)
            dis = np.sqrt(np.sum(np.square(self.train_x - test_x[i]), axis=1))
            k_min_labels = self.train_y[np.argsort(dis)[:self.n_neighbours]]
            # 分类，使用投票原则
            if self.classification is True:
                k_labels_count = np.bincount(k_min_labels)
                pre[i] = np.argmax(k_labels_count)
            else:         # 回归，采用平均值原则
                pre[i] = np.mean(k_min_labels)
        return pre
```

```python
def load_data(path = "../data/mnist", kind = 'train', one_hot = True, normals = True):
    images, labels = load_mnist(path, kind=kind)
    if normals == True:
        images = images / 255.0
    if one_hot == True:
        labels_onehot = np.zeros((labels.shape[0], np.unique(labels).shape[0]))
        for i in range(labels_onehot.shape[0]):
            labels_onehot[i][labels[i]] = 1
        labels = labels_onehot
    return (images, labels)
```

```python
train_x, train_y = load_data(kind='train', one_hot=False)
```

```python
less_test_x = test_x[mask]
less_test_y = test_y[mask]
```

## knn的决策边界

```python
def plot_decision_boundary(model, grid_x, grid_y, test_x):
    cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
    cmap_bold  = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])
    train_x = model.train_x
    train_y = model.train_y
    
    x_0_min, x_0_max = train_x[:, 0].min() - 1, train_x[:, 0].max() + 1
    x_1_min, x_1_max = train_x[:, 1].min() - 1, train_x[:, 1].max() + 1
    
    pre = model.predict(test_x)
    pre = pre.reshape(grid_x.shape)
    
    plt.figure(figsize=(12, 7))
    plt.pcolormesh(grid_x, grid_y, pre, cmap=cmap_light)
    plt.scatter(train_x[:, 0], train_x[:, 1], c=train_y, cmap=cmap_bold, edgecolors='k', 
                s=20)

    plt.xlim((x_0_min+1, x_0_max-1))
    plt.ylim((x_1_min+1, x_1_max-1))

    plt.title("3-Class classification (k = %i)" % (model.n_neighbours))
```

```python
def load_data_iris():
    iris = load_iris()
    X = iris.data
    y = iris.target
    x_0_min, x_0_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x_1_min, x_1_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    density = 0.02  # 步长（表示数据的分布密度）
    
    new_x, new_y = np.meshgrid(np.arange(start = x_0_min, stop = x_0_max, step = density),
    np.arange(start = x_1_min, stop = x_1_max, step = density))
    
    test_x = np.c_[new_x.ravel(), new_y.ravel()]
    
    return X[:, :2], y, new_x, new_y, test_x
```

```python
train_x, train_y, grid_x, grid_y, test_x = load_data_iris()
```

```python
for k in [1, 2, 5, 10]:
    model = Knn(n_neighbours = k)
    model.train(train_x, train_y)
    plot_decision_boundary(model, grid_x, grid_y, test_x)
```

### knn做回归任务 

```python
def load_data_for_regression():
    np.random.seed(78)
    X = np.sort(5 * np.random.rand(40, 1), axis=0)
    T = np.linspace(0, 5, 500)[:, np.newaxis]
    y = np.sin(X).ravel()
    y[::5] += 1 * (0.5 - np.random.rand(8))
    return X, y, T
```

```python
def regression(train_x, train_y, test_x):
    plt.figure(figsize=(12, 7))
    for i, k in enumerate([1, 2, 5, 10]):
        model = Knn(n_neighbours = k, classification=False) # 训练一个回归器
        model.train(train_x, train_y)
        pre_y = model.predict(test_x)
        
        plt.subplot(2, 2, i+1)
        plt.scatter(train_x, train_y, c='k', label='data')
        plt.plot(test_x, pre_y, c='g', label='prediction')
        
        plt.axis('tight')
        plt.legend()
        plt.title("KNeighborsRegressor (neighbours = %i)" % k)
```

```python
train_x, train_y, test_x = load_data_for_regression()
```

```python
regression(train_x, train_y, test_x)
```
